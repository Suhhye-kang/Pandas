```python
raw=korea.loc[:,['2','4','8','19','99','101','106','108','114','120','121','122','123','142']]
raw
```

```python
raw.columns=['가구별일련번호','가구원수','맞벌이가구','가구주_직업코드','거처구분','입주형태', '주택소유구분','소득5분위코드','소득10분위코드','소득구간', '가계지출', '소비지출', '식료품','주류']
```

```python
raw
```

```python
raw.groupby('소득구간').size()
```

```python
raw['소득']=raw['소득구간']*1000000
```

```python
raw
```

```python
raw['소득대비지출비중']=raw['소비지출']/raw['소득']
raw['소득대비식비비중']=raw['식료품']/raw['소득']
raw['소득대비주류비비중']=raw['주류']/raw['소득']
```

```python
raw
```

```python
plot=raw.groupby(['소득구간']).mean()
```

```python
plot
```
