```python
import pandas as pd
```

```python
all=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_all.xlsx")
```

```python
all
```

```python
all['size_range']=pd.cut(all['size'],[0, 65, 98, 132, 10000], labels=['20-','20-30','30-40','40+'], duplicates='raise')
```

```python
all.loc[all['size']==66]
```

```python
all['age']=2020-all['year_built']
```

```python
all['status']=pd.cut(all['age'],[0, 5, 20,1000], labels=['신축','중축','구축'], duplicates='raise')
```

```python
all
```

```python
status=all.groupby(['status','size_range','yy'])['price_sold'].mean()
```

```python
status=status.to_frame()
```

```python
status=status.reset_index()
```

```python
status=pd.pivot_table(status,index=['status','size_range'],columns='yy',values='price_sold')
```

```python
status.to_excel('/Users/kang/Pandas/Real/status_apt.xlsx')
```
