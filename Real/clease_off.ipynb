```python
import pandas as pd
```

```python
off_sell_19=pd.read_excel("/Users/kang/Pandas/Real/off_sell_2019.xlsx")
off_sell_18=pd.read_excel("/Users/kang/Pandas/Real/off_sell_2018.xlsx")
off_sell_17=pd.read_excel("/Users/kang/Pandas/Real/off_sell_2017.xlsx")
off_sell_16=pd.read_excel("/Users/kang/Pandas/Real/off_sell_2016.xlsx")
off_sell_15=pd.read_excel("/Users/kang/Pandas/Real/off_sell_2015.xlsx")
off_sell_14=pd.read_excel("/Users/kang/Pandas/Real/off_sell_2014.xlsx")
off_sell_13=pd.read_excel("/Users/kang/Pandas/Real/off_sell_2013.xlsx")
off_sell_12=pd.read_excel("/Users/kang/Pandas/Real/off_sell_2012.xlsx")
off_sell_11=pd.read_excel("/Users/kang/Pandas/Real/off_sell_2011.xlsx")
off_sell_10=pd.read_excel("/Users/kang/Pandas/Real/off_sell_2010.xlsx")
```

```python
off_sell_2010_2019=pd.concat([off_sell_19,off_sell_18,off_sell_17,off_sell_16,off_sell_15,off_sell_14,off_sell_13,off_sell_12,off_sell_11,off_sell_10])
off_sell_2010_2019
```

```python
off_sell_2010_2019.columns=['citytownvillage','zipcode','zipcode_main','zipcode_sub','name','size','yymm','dd','price_sold','floor','year_built','address']
```

```python
off_sell_2010_2019['citytownvillage']=off_sell_2010_2019['citytownvillage'].astype(str)
off_sell_2010_2019['name']=off_sell_2010_2019['name'].astype(str)
off_sell_2010_2019['address']=off_sell_2010_2019['address'].astype(str)
off_sell_2010_2019['size']=off_sell_2010_2019['size'].astype(int)
off_sell_2010_2019['price_sold']=off_sell_2010_2019['price_sold'].astype(str).str.replace(',','')
off_sell_2010_2019['price_sold']=off_sell_2010_2019['price_sold'].astype(int)
off_sell_2010_2019['city']=off_sell_2010_2019['citytownvillage'].str.split(' ').str[0]
off_sell_2010_2019['town']=off_sell_2010_2019['citytownvillage'].str.split(' ').str[1]
off_sell_2010_2019['village']=off_sell_2010_2019['citytownvillage'].str.split(' ').str[2]
off_sell_2010_2019['yy']=off_sell_2010_2019['yymm'].astype(str).str.slice(0,4)
off_sell_2010_2019['mm']=off_sell_2010_2019['yymm'].astype(str).str.slice(4,6)
off_sell_2010_2019['yy']=off_sell_2010_2019['yy'].astype(int)
off_sell_2010_2019['mm']=off_sell_2010_2019['mm'].astype(int)
```

```python
del off_sell_2010_2019['citytownvillage']
del off_sell_2010_2019['yymm']
del off_sell_2010_2019['zipcode_main']
del off_sell_2010_2019['zipcode_sub']
```

```python
neworder = ['city','town','village','name', 'size','price_sold','floor','year_built','yy','mm','dd','zipcode','address']
off_sell_2010_2019=off_sell_2010_2019.reindex(columns=neworder)
```

```python
off_sell_2010_2019.to_excel('/Users/kang/Pandas/Real/off_sell_all.xlsx')
```
