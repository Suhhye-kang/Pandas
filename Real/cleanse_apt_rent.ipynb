```python
import pandas as pd
```

```python
apt_rent_19=pd.read_excel("/Users/kang/Pandas/Real/apt_rent_2019.xlsx")
apt_rent_18=pd.read_excel("/Users/kang/Pandas/Real/apt_rent_2018.xlsx")
apt_rent_17=pd.read_excel("/Users/kang/Pandas/Real/apt_rent_2017.xlsx")
apt_rent_16=pd.read_excel("/Users/kang/Pandas/Real/apt_rent_2016.xlsx")
apt_rent_15=pd.read_excel("/Users/kang/Pandas/Real/apt_rent_2015.xlsx")
apt_rent_14=pd.read_excel("/Users/kang/Pandas/Real/apt_rent_2014.xlsx")
apt_rent_13=pd.read_excel("/Users/kang/Pandas/Real/apt_rent_2013.xlsx")
apt_rent_12=pd.read_excel("/Users/kang/Pandas/Real/apt_rent_2012.xlsx")
apt_rent_11=pd.read_excel("/Users/kang/Pandas/Real/apt_rent_2011.xlsx")
apt_rent_10=pd.read_excel("/Users/kang/Pandas/Real/apt_rent_2010.xlsx")
```

```python
apt_rent_2010_2019=pd.concat([apt_rent_19,apt_rent_18,apt_rent_17,apt_rent_16,apt_rent_15,apt_rent_14,apt_rent_13,apt_rent_12,apt_rent_11,apt_rent_10])
```

```python
apt_rent_2010_2019.columns=['citytownvillage','zipcode','zipcode_main','zipcode_sub','name','category','size','yymm','dd','deposit','monthlyrent','floor','year_built','address']
```

```python
apt_rent_2010_2019
```

```python
apt_rent_2010_2019['citytownvillage']=apt_rent_2010_2019['citytownvillage'].astype(str)
apt_rent_2010_2019['name']=apt_rent_2010_2019['name'].astype(str)
apt_rent_2010_2019['deposit']=apt_rent_2010_2019['deposit'].astype(str).str.replace(',','')
apt_rent_2010_2019['monthlyrent']=apt_rent_2010_2019['monthlyrent'].astype(str).str.replace(',','')
```

```python

apt_rent_2010_2019['deposit']=apt_rent_2010_2019['deposit'].astype(int)
apt_rent_2010_2019['monthlyrent']=apt_rent_2010_2019['monthlyrent'].astype(int)
apt_rent_2010_2019['city']=apt_rent_2010_2019['citytownvillage'].str.split(' ').str[0]
apt_rent_2010_2019['town']=apt_rent_2010_2019['citytownvillage'].str.split(' ').str[1]
apt_rent_2010_2019['village']=apt_rent_2010_2019['citytownvillage'].str.split(' ').str[2]
apt_rent_2010_2019['yy']=apt_rent_2010_2019['yymm'].astype(str).str.slice(0,4)
apt_rent_2010_2019['mm']=apt_rent_2010_2019['yymm'].astype(str).str.slice(4,6)
apt_rent_2010_2019['yy']=apt_rent_2010_2019['yy'].astype(int)
apt_rent_2010_2019['mm']=apt_rent_2010_2019['mm'].astype(int)
```

```python
del apt_rent_2010_2019['citytownvillage']
del apt_rent_2010_2019['yymm']
del apt_rent_2010_2019['zipcode_main']
del apt_rent_2010_2019['zipcode_sub']
```
