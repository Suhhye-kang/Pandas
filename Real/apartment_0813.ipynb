```python
import pandas as pd
```

```python
apt_sell_19=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_2019.xlsx")
apt_sell_18=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_2018.xlsx")
apt_sell_17=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_2017.xlsx")
apt_sell_16=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_2016.xlsx")
apt_sell_15=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_2015.xlsx")
apt_sell_14=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_2014.xlsx")
apt_sell_13=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_2013.xlsx")
apt_sell_12=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_2012.xlsx")
apt_sell_11=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_2011.xlsx")
apt_sell_10=pd.read_excel("/Users/kang/Pandas/Real/apt_sell_2010.xlsx")
```

```python
apt_sell_2010_2019=pd.concat([apt_sell_19,apt_sell_18,apt_sell_17,apt_sell_16,apt_sell_15,apt_sell_14,apt_sell_13,apt_sell_12,apt_sell_11,apt_sell_10])
apt_sell_2010_2019
```

```python
apt_sell_2010_2019.columns=['citytownvillage','zipcode','zipcode_main','zipcode_sub','name','size','yymm','dd','price_sold','floor','year_built','address']
```

```python
apt_sell_2010_2019.info()
```

```python
apt_sell_2010_2019['citytownvillage']=apt_sell_2010_2019['citytownvillage'].astype(str)
apt_sell_2010_2019['name']=apt_sell_2010_2019['name'].astype(str)
apt_sell_2010_2019['address']=apt_sell_2010_2019['address'].astype(str)
apt_sell_2010_2019['size']=apt_sell_2010_2019['size'].astype(int)
apt_sell_2010_2019['price_sold']=apt_sell_2010_2019['price_sold'].astype(str).str.replace(',','')
apt_sell_2010_2019['price_sold']=apt_sell_2010_2019['price_sold'].astype(int)
apt_sell_2010_2019['city']=apt_sell_2010_2019['citytownvillage'].str.split(' ').str[0]
apt_sell_2010_2019['town']=apt_sell_2010_2019['citytownvillage'].str.split(' ').str[1]
apt_sell_2010_2019['village']=apt_sell_2010_2019['citytownvillage'].str.split(' ').str[2]
apt_sell_2010_2019['yy']=apt_sell_2010_2019['yymm'].astype(str).str.slice(0,4)
apt_sell_2010_2019['mm']=apt_sell_2010_2019['yymm'].astype(str).str.slice(4,6)
apt_sell_2010_2019['yy']=apt_sell_2010_2019['yy'].astype(int)
apt_sell_2010_2019['mm']=apt_sell_2010_2019['mm'].astype(int)
```

```python
del apt_sell_2010_2019['citytownvillage']
del apt_sell_2010_2019['yymm']
del apt_sell_2010_2019['zipcode_main']
del apt_sell_2010_2019['zipcode_sub']
```

```python
neworder = ['city','town','village','name', 'size','price_sold','floor','year_built','yy','mm','dd','zipcode','address']
apt_sell_2010_2019=apt_sell_2010_2019.reindex(columns=neworder)
```
