```python
all20['시군구'].astype(str)
```

```python
all20['광역시']=all20['시군구'].str.split(' ').str[0]
all20['자치구']=all20['시군구'].str.split(' ').str[1]
all20['동']=all20['시군구'].astype(str).str.split(' ').str[2]
```

```python
all20['year']=all20['계약년월'].astype(str).str.slice(0,4)
all20['month']=all20['계약년월'].astype(str).str.slice(4,6)
```

```python
all20['매매가']=all20['매매가'].str.replace(',','')
all20['보증금']=all20['보증금'].str.replace(',','')
all20['월세']=all20['월세'].str.replace(',','')
```

```python
all20['매매가'].fillna(0,inplace=True)
all20['보증금'].fillna(0,inplace=True)
all20['월세'].fillna(0,inplace=True)
```

```python
all20['매매가']=all20['매매가'].astype(str).astype(int)
all20['보증금']=all20['보증금'].astype(str).astype(int)
all20['보증금']=all20['월세'].astype(str).astype(int)
```

```python
all20.info()
```

```python
sell20=all20.loc[all20['구분']=='매매']
```

```python
sell_month=sell20.groupby(['자치구','계약년월'])['매매가'].mean()
sell_month=sell_month.to_frame().reset_index()
```

```python
sell_month
```

```python
sell_month=sell_month.pivot(index='계약년월',columns='자치구',values='매매가')
```
