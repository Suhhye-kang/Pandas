# 다변량 그래프 그리기

```python
def recode_sex(sex):
    if sex=='Female':
        return 0
    else:
        return 1
```

```python
tips['sex_color']=tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    alpha=0.5
)

axes1.set_title('Total Bill vs. Tips Colored by Sex and Sized by Size')
axes1.set_xlabel('total bill')
axes1.set_ylabel('tips')
```

```python
gapminder=pd.read_csv('C:\\Github\\Pandas\\doit\\data\\gapminder.tsv',sep='\t')
gapminder
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
gdp_by_year = gapminder.groupby('year')['gdpPercap'].mean()
gdp_by_year.plot()
```

```python
gdp_by_year=gapminder.groupby('year',as_index=False)['gdpPercap'].mean()
gdp_by_year
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.scatter(x=gdp_by_year['year'],y=gdp_by_year['gdpPercap'])
axes1.set_title('GDP Growth by Year')
```

```python
gdp_by_continent=gapminder.groupby(['continent','year'],as_index=False)['gdpPercap'].mean()
gdp_by_continent
```

```python
def recode_continent(continent):
    if continent == 'Oceania':
        return 0
    elif continent == 'Europe':
        return 1
    elif continent == 'Asia':
        return 2
    elif continent == 'Americas':
        return 3
    else: 
        return 4
```

```python
gdp_by_continent['continent_color']=gdp_by_continent['continent'].apply(recode_continent)
```

```python
fig=plt.figure()
axes1=fig.add_subplot(1,1,1)
axes1.scatter(x=gdp_by_continent['year'],y=gdp_by_continent['gdpPercap'],
              c=gdp_by_continent['continent_color']
)
axes1.set_title('GDP by Continent')
axes1.set_xlabel('year')
axes1.set_ylabel('GDP')
```

```python
boxplot=plt.figure()
axes1=boxplot.add_subplot(1,1,1)
axes1.boxplot([gapminder[gapminder['continent']=='Africa']['pop'],
              gapminder[gapminder['continent']=='Asia']['pop'],
              gapminder[gapminder['continent']=='Europe']['pop'],
              gapminder[gapminder['continent']=='Oceania']['pop'],
              gapminder[gapminder['continent']=='Americas']['pop']]
)
```

# 데이터프레임과 시리즈로 그래프 그리기

```python
import seaborn as sns
tips=sns.load_dataset("tips")
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Historgram with Density Plot')
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],kde=False)
ax.set_title("Total Bill Histogram")
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'],hist=False)
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
ax=plt.subplots()
ax=sns.distplot(tips['total_bill'],rug=True)
ax.set_title("Total Bill Histogram with Density and Rug Plot")
ax.set_xlabel('Total Bill')
```

```python
ax=plt.subplots()
ax=sns.countplot('day',data=tips)
ax.set_title("Count of days")
ax.set_xlabel('Day')
ax.set_ylabel("Frequency")
```

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips)
ax.set_title("Scatter plot of total bills and tip")
ax.set_xlabel('total bill')
ax.set_ylabel('tips')
```

```python
ax=plt.subplots()
ax=sns.regplot(x='total_bill',y='tip',data=tips,fit_reg=False)
ax.set_title("Scatter Plot of Total Bills and Tip")
ax.set_xlabel('Total Bill')
ax.set_ylabel('tips')
```

```python
joint=sns.jointplot(x='total_bill',y='tip',data=tips)
joint.set_axis_labels(xlabel='total bill',ylabel='tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip',fontsize=10,y=1.03)
```

```python
hexbin=sns.jointplot(x="total_bill",y='tip',data=tips,kind="hex")
hexbin.set_axis_labels(xlabel="total bill",ylabel='tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill',fontsize=15, y=1.03)
```

```python
ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'],
               data2=tips['tip'],
               shade=True)
ax.set_title("Kernel Density Estimation of Total Bill and Tip")
ax.set_xlabel("total bill")
ax.set_ylabel('tip')
```

```python
ax=plt.subplots()
ax=sns.barplot(x='time',y='total_bill',data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('time')
ax.set_ylabel('total bill')
```

```python
ax=plt.subplots()
ax=sns.boxplot(x='time',y='total_bill',data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('time')
ax.set_ylabel('total bill')
```

```python
ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',data=tips)
ax.set_title("Violin Plot of total bill by time of day")
ax.set_xlabel('time')
ax.set_ylabel('total bill')
```

```python
fig=sns.pairplot(tips)
```

```python
pair_grid=sns.PairGrid(tips)
pair_grid=pair_grid.map_upper(sns.regplot)
pair_grid=pair_grid.map_lower(sns.kdeplot)
pair_grid=pair_grid.map_diag(sns.distplot,rug=True)
plt.show()
```
