# 열 이름 나누고 데이터 프레임에 추가하기

```python
weather=pd.read_csv("C:\\Github\\Pandas\\doit\\data\\weather.csv")
weather
```

```python
weather_melt=pd.melt(weather,id_vars=['id','year','month','element'],var_name='day',value_name='temperature')
weather_melt
```

```python
weather_tidy=weather_melt.pivot_table(
index=['id','year','month','day'],
    columns='element',
    values='temperature'
)
weather_tidy
```

```python
weather_tidy_flat=weather_tidy.reset_index()
weather_tidy_flat
```

```python
day_edit = weather_tidy_flat.day.str[1:]
print(day_edit)
```

```python
weather_tidy_flat['day_edit']=day_edit
weather_tidy_flat
```

```python
weather_tidy_edit=weather_tidy_flat.drop(['day'],axis=1)
weather_tidy_edit
```

```python
wy=pd.to_datetime(weather_tidy_edit['year'],format='%Y')
wy
```

# 뉴욕 택시 데이터 준비

```python
import os 
import urllib.request

# 네트워크 상태에 따라 5 ~ 15분이 소요됩니다.
with open('../data/raw_data_urls.txt', 'r') as data_urls:
    for line, url in enumerate(data_urls):
        if line == 5:
            break 
        fn = url.split('/')[-1].strip()
        fp = os.path.join('', '../data', fn)
        print(url)
        print(fp)
        urllib.request.urlretrieve(url, fp)
```

```python
import glob 
nyc_taxi_data = glob.glob('../data/fhv_*') 
print(nyc_taxi_data)
```

```python
taxi1=pd.read_csv(nyc_taxi_data[0])
taxi2=pd.read_csv(nyc_taxi_data[1])
taxi3=pd.read_csv(nyc_taxi_data[2])
taxi4=pd.read_csv(nyc_taxi_data[3])
taxi5=pd.read_csv(nyc_taxi_data[4])
```

```python
taxi1
```

```python
print(taxi1.shape)
print(taxi2.shape)
print(taxi3.shape)
print(taxi4.shape)
print(taxi5.shape)
```

```python
taxi=pd.concat([taxi1,taxi2,taxi3,taxi4,taxi5])
print(taxi.shape)
```

# 반복문으로 데이터 준비하기

```python
list_taxi_df=[]

for csv_filename in nyc_taxi_data:
    # print(csv_filename)
    df=pd.read_csv(csv_filename)
    list_taxi_df.append(df)
    
print(len(list_taxi_df))
```
