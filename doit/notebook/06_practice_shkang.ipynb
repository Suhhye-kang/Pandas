# 데이터를 입력할 때 누락값이 발생하는 경우

```python
gapminder=pd.read_csv("C:\\Github\\Pandas\\doit\\data\\gapminder.tsv",sep='\t')
```

```python
life_exp=gapminder.groupby(['year'])['lifeExp'].mean()
print(life_exp)
```

# 인덱스를 다시 만들 때 누락값이 발생하는 경우

```python
y2000=life_exp[life_exp.index>2000]
print(y2000)
```

# 누락값의 개수 구하기

```python
ebola=pd.read_csv("C:\\Github\\Pandas\\doit\\data\\country_timeseries.csv")
ebola
```

```python
print(ebola.count())
```

```python
print(ebola.shape[0])
```

```python
pd.isnull(ebola)
```

```python
num_rows=ebola.shape[0]
num_missing=num_rows-ebola.count()
print(num_missing)
```

```python
import numpy as np
print(np.count_nonzero(ebola.isnull()))
```

```python
print(np.count_nonzero(ebola['Cases_Guinea'].isnull()))
```

```python
print(ebola.Cases_Guinea.value_counts(dropna=False))
```

# 누락값을 다른 값으로 변경하기

```python
print(ebola.fillna(0).iloc[0:10,0:5])
```

```python
print(ebola.fillna(method='ffill').iloc[0:10,0:5])
```

```python
print(ebola.fillna(method='bfill').iloc[0:10,0:5])
```

```python
print(ebola.interpolate().iloc[0:10,0:5])
```
